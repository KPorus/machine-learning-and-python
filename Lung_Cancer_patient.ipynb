```python
import pandas as pd
```

```python
df = pd.read_csv("cancer patient data sets.csv")
df.head()
```

```python
df.drop(["Patient Id","index"], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Level"] = le.fit_transform(df["Level"])
```

```python
df.head()
```

```python
df.corr()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

sns.pairplot(df,vars=['Age', 'Gender', 'Air Pollution', 'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 'Obesity', 'Smoking', 'Passive Smoker','Coughing of Blood', 'Chest Pain', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing'])
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[]
df1 = df.drop(["Level","Alcohol use","Dust Allergy","Balanced Diet"], axis=1)
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,15),wcss,marker='o',linestyle='--')
plt.grid()
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
kmeans.fit(df1)
```

```python
df['cluster'] = kmeans.predict(df1)
```

```python
cluster_means = df.groupby('cluster').mean()
print(cluster_means)
```

# using regrassion methods

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df1,df["Level"],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
x_train = poly.fit_transform(x_train)
x_test = poly.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
res = lr.predict(x_test)
res = pd.DataFrame(res,columns=["predicted level"])
res.head(100)
```

# Summary

We applied the **K-means** clustering algorithm for grouping based on levels and the **Polynomial** algorithm for predicting levels.

From our analysis, we found that the clustering results were not accurate, while the regression model achieved a **99%** accuracy in predicting lung cancer levels.

Based on these findings, we concluded that using the **Polynomial algorithm** in conjunction with **Logistic Regression** would be the most effective approach.
